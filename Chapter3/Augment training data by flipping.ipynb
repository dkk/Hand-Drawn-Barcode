```python
BARS_FOLDER = "../rotated_barcodes/bars/"
OUTPUT_FOLDER = "../rotated_barcodes/bars/"
```

```python
from os import listdir
from os.path import isfile, join

paths = [BARS_FOLDER + f for f in listdir(BARS_FOLDER) if isfile(join(BARS_FOLDER, f))]
```

```python
import pandas as pd

df = pd.DataFrame({'path': paths})
df["dec"] = df.apply(lambda x: int(x['path'].split("/")[-1].split("_")[0]), axis=1)
df["bin"] = df.apply(lambda x: bin(x['dec']), axis=1)
df["class"] = df.apply(lambda x: [i for i in reversed(range(20)) if (x['dec'] & 1 << i) != 0], axis=1)
    
print(df.shape)
df.head()
```

```python
from PIL import Image
from PIL import ImageOps
```

```python
h_m = {0:1, 1:0, 2:3, 3:2, 
       4:5, 5:4, 6:7, 7:6, 
       8:9, 9:8, 10:11, 11:10, 
       12:13, 13:12, 14:15, 15:14, 
       16:17, 17:16, 18:19, 19:18}

v_m = {0:19, 19:0, 3:16, 16:3,
      1:18, 18:1, 2:17, 17:2,
      4:15, 15:4, 5:14, 14:5,
      7:12, 12:7, 6:13, 13:6,
      8:11, 11:8, 9:10, 10:9}


def h_mirror_code(code):
    mirror_code = 0
    for i in range(20):
        mirror_code += (1 if code & 1 << h_m[i] != 0 else 0) << i
    
    return mirror_code    

def v_mirror_code(code):
    mirror_code = 0
    for i in range(20):
        mirror_code += (1 if code & 1 << v_m[i] != 0 else 0) << i
    
    return mirror_code 

def hv_mirror_code(code):
    mirror_code = 0
    for i in range(20):
        mirror_code += (1 if code & 1 << h_m[v_m[i]] != 0 else 0) << i
    
    return mirror_code 
```

```python
# generate horizontally flipped images
for path, dec in zip(df["path"], df["dec"]):
    print(path, dec)
    im = Image.open(path)
    im = ImageOps.mirror(im)
    new_code = h_mirror_code(dec)
    print(new_code)
    im.save("{}/{}_h_{}.png".format(OUTPUT_FOLDER, new_code, dec))
```

```python
# generate vertically flipped images
for path, dec in zip(df["path"], df["dec"]):
    print(path, dec)
    im = Image.open(path)
    im = ImageOps.flip(im)
    new_code = v_mirror_code(dec)
    print(new_code)
    im.save("{}/{}_v_{}.png".format(OUTPUT_FOLDER, new_code, dec))
```

```python
# generate horizontally and vertically flipped images
for path, dec in zip(df["path"], df["dec"]):
    print(path, dec)
    im = Image.open(path)
    im = ImageOps.mirror(im)
    im = ImageOps.flip(im)
    new_code = hv_mirror_code(dec)
    print(new_code)
    im.save("{}/{}_hv_{}.png".format(OUTPUT_FOLDER, new_code, dec))
```
